```python
params = {
    "neuron": {
        "type": "SRMALPHA",
        "theta": 5, # 10
        "tauSr": 10.0,
        "tauRef": 2.0,
        "scaleRef": 2,
        "tauRho": 1,
        "scaleRho": 1,
    },
    "simulation": {"Ts": 1.0, "tSample": 75, "nSample": 1},
    "training": {
        "error": {
            "type": "SpikeTime",  # "NumSpikes" or "ProbSpikes"
            "probSlidingWin": 20,  # only valid for ProbSpikes
            "tgtSpikeRegion": {  # valid for NumSpikes and ProbSpikes
                "start": 0,
                "stop": 75,
            },
            "tgtSpikeCount": {True: 60, False: 10},
        }
    },
}
```

```python
class SlayerMLP(torch.nn.Module):
    def __init__(self, params, input_size, hidden_size1, hidden_size2):
        super(SlayerMLP, self).__init__()
        self.slayer = snn.layer(params["neuron"], params["simulation"])
        self.fc1 = self.slayer.dense(input_size, hidden_size1)
        self.fc2 = self.slayer.dense(hidden_size1, hidden_size2)
        
        self.fc_decoder1 = self.slayer.dense(hidden_size2, hidden_size1)
        self.fc_decoder2 = self.slayer.dense(hidden_size1, input_size)

    def get_spike(self, inp):
        return self.slayer.spike(inp)
    
    def forward(self, spike_input):
        spike_1 = self.slayer.spike(self.slayer.psp(self.fc1(spike_input)))
        spike_2 = self.slayer.spike(self.slayer.psp(self.fc2(spike_1)))
        
        spike_hat1 = self.slayer.spike(self.slayer.psp(self.fc_decoder1(spike_2)))
        spike_hat2 = self.slayer.spike(self.slayer.psp(self.fc_decoder2(spike_hat1)))
        
        return spike_2, spike_hat2
```

```python
icub_dir = 'models_and_stats/semisupervised_both_0_icub.pt'
bio_dir = 'models_and_stats/semisupervised_both_0_bio.pt'
```

```python
device = torch.device("cuda:2")
net_icub = SlayerMLP(params, 60, 50, 50).to(device)
net_bio = SlayerMLP(params, 19, 50, 50).to(device)

net_icub.load_state_dict(torch.load(icub_dir))
net_bio.load_state_dict(torch.load(bio_dir))

net_icub.eval()
net_bio.eval()

error = snn.loss(params).to(device)
```

```python
def get_metrics(lloader, _net, bio=True):
    # Draw representations
    _net.eval()
    big_list = []
    label_list=[]
    with torch.no_grad():
        for i, (tact_icub, tact_bio, target, label) in enumerate(lloader):
            if bio:
                tact_bio = tact_bio.to(device)
                tact_bio = tact_bio[...,::2]
                tact = tact_bio
            else:
                tact_icub = tact_icub.to(device)
                tact = tact_icub
            
            target = target.to(device)
            tact = _net.get_spike(tact)
            rep, _ = _net.forward(tact)
            big_list.append(rep)
            label_list.append(label)

    test_representation = torch.cat(big_list,0)
    label_all = torch.cat(label_list, 0)
    label_all = label_all.cpu().numpy()
    dist_mat = torch.zeros([len(lloader.dataset),len(lloader.dataset)])
    for i in range(len(lloader.dataset)):
        for j in range(len(lloader.dataset)):
            if i == j:
                dist_mat[i, j] = 0.0
            else:
                dist_mat[i,j] = error.spikeTime(test_representation[i], test_representation[j])

    dist_mat = dist_mat.detach().cpu().numpy()
    
    return dist_mat, label_all
```

```python
def get_metrics2(lloader, _net_icub, _net_bio):
    # Draw representations
    _net_icub.eval()
    _net_bio.eval()
    big_list_icub = []
    big_list_bio = []
    label_list=[]
    with torch.no_grad():
        for i, (tact_icub, tact_bio, target, label) in enumerate(lloader):
            tact_bio = tact_bio.to(device)
            tact_bio = tact_bio[...,::2]
            tact_icub = tact_icub.to(device)
            target = target.to(device)
            
            tact_icub = _net_icub.get_spike(tact_icub)
            tact_bio = _net_bio.get_spike(tact_bio)
            
            rep_icub, _ = _net_icub.forward(tact_icub)
            rep_bio, _ = _net_bio.forward(tact_bio)
            
            #big_list.append([rep_icub, rep_bio])
            big_list_icub.append(rep_icub)
            big_list_bio.append(rep_bio)
            label_list.append(label)

    test_representation_icub = torch.cat(big_list_icub,0)
    test_representation_bio = torch.cat(big_list_bio,0)
    label_all = torch.cat(label_list, 0)
    label_all = label_all.cpu().numpy()
    dist_mat = torch.zeros([len(lloader.dataset),len(lloader.dataset)])
    for i in range(len(lloader.dataset)):
        for j in range(len(lloader.dataset)):
            dist_mat[i,j] = error.spikeTime(test_representation_icub[i], test_representation_bio[j])

    dist_mat = dist_mat.detach().cpu().numpy()
    
    return dist_mat, label_all
```

### Train simple classifier

```python
class Classifier(torch.nn.Module):
    def __init__(self, params, input_size, hidden_size, output_size):
        super(Classifier, self).__init__()
        self.slayer = snn.layer(params["neuron"], params["simulation"])
        self.fc1 = self.slayer.dense(input_size, hidden_size)
        self.fc2 = self.slayer.dense(hidden_size, output_size)

    def get_spike(self, inp):
        return self.slayer.spike(inp)
    
    def forward(self, spike_input):
        spike_1 = self.slayer.spike(self.slayer.psp(self.fc1(spike_input)))
        spike_2 = self.slayer.spike(self.slayer.psp(self.fc2(spike_1)))
        
        return spike_2
```

```python
params2 = {
    "neuron": {
        "type": "SRMALPHA",
        "theta": 5, # 10
        "tauSr": 10.0,
        "tauRef": 2.0,
        "scaleRef": 2,
        "tauRho": 1,
        "scaleRho": 1,
    },
    "simulation": {"Ts": 1.0, "tSample": 75, "nSample": 1},
    "training": {
        "error": {
            "type": "NumSpikes",  # "NumSpikes" or "ProbSpikes"
            "probSlidingWin": 20,  # only valid for ProbSpikes
            "tgtSpikeRegion": {  # valid for NumSpikes and ProbSpikes
                "start": 0,
                "stop": 75,
            },
            "tgtSpikeCount": {True: 60, False: 10},
        }
    },
}
```

```python
net_icub_clr = Classifier(params2, 50, 128, 20).to(device)
net_bio_clr = Classifier(params2, 50, 128, 20).to(device)
error_icub = snn.loss(params2).to(device)
error_bio = snn.loss(params2).to(device)
optimizer_icub = torch.optim.RMSprop(net_icub_clr.parameters(), lr=0.01, weight_decay=0.5)
optimizer_bio = torch.optim.RMSprop(net_bio_clr.parameters(), lr=0.01, weight_decay=0.5)
```

```python
for param in net_icub.parameters():
    param.requires_grad = False
for param in net_bio.parameters():
    param.requires_grad = False
```

```python
save_dir_icub = 'models_and_stats/shared_icub_0.pt'
save_dir_bio = 'models_and_stats/shared_bio_0.pt'
```

```python
train_icub_losses = []
train_bio_losses = []
train_icub_accs = []
train_bio_accs = []

val_icub_losses = []
val_bio_losses = []
val_icub_accs = []
val_bio_accs = []

max_icub_val_acc = 0
max_bio_val_acc = 0

for epoch in range(3001):
    net_icub_clr.train()
    net_bio_clr.train()
    loss_train_icub = 0
    loss_train_bio = 0
    correct_icub = 0
    correct_bio = 0
    for i, (tact_icub, tact_bio,  target, label) in enumerate(train_loader):
        
        tact_bio = tact_bio.to(device)
        tact_bio = tact_bio[...,::2]
        tact_icub = tact_icub.to(device)
        
        target = target.to(device)
        tact_bio = net_bio.get_spike(tact_bio)
        tact_icub = net_icub.get_spike(tact_icub)
        
        with torch.no_grad():
            icub_rep, _ = net_icub.forward(tact_icub)
            bio_rep, _ = net_bio.forward(tact_bio)
            
        output_icub = net_icub_clr.forward(icub_rep)
        output_bio = net_bio_clr.forward(bio_rep)
            
        correct_icub += torch.sum(snn.predict.getClass(output_icub) == label).data.item()
        correct_bio += torch.sum(snn.predict.getClass(output_bio) == label).data.item()
                
#         print(icub_rep.shape, target.shape)
        loss_icub = error_icub.numSpikes(output_icub, target)
        loss_bio = error_bio.numSpikes(output_bio, target)
        
        loss_train_icub += loss_icub.item()
        loss_train_bio += loss_bio.item()
        
        optimizer_icub.zero_grad()
        optimizer_bio.zero_grad()
        loss_icub.backward()
        loss_bio.backward()
        optimizer_icub.step()
        optimizer_bio.step()
                
    if epoch%10 == 0:
        print('Epoch: ', epoch, ' --------------------------')
        print('Train loss (icub, bio):', 
              loss_train_icub/len(train_dataset),
              loss_train_bio/len(train_dataset))
        print('Train acc (icub, bio):', 
              correct_icub/len(train_dataset),
              correct_bio/len(train_dataset))
    train_icub_losses.append(loss_train_icub/len(train_dataset))
    train_bio_losses.append(loss_train_bio/len(train_dataset))
    train_icub_accs.append(correct_icub/len(train_dataset))
    train_bio_accs.append(correct_bio/len(train_dataset))
    
        
    net_icub_clr.eval()
    net_bio_clr.eval()
    loss_train_icub = 0
    loss_train_bio = 0
    correct_icub = 0
    correct_bio = 0
    with torch.no_grad():
        for i, (tact_icub, tact_bio,  target, label) in enumerate(val_loader):

            tact_bio = tact_bio.to(device)
            tact_bio = tact_bio[...,::2]
            tact_icub = tact_icub.to(device)

            target = target.to(device)
            tact_bio = net_bio.get_spike(tact_bio)
            tact_icub = net_icub.get_spike(tact_icub)

            with torch.no_grad():
                icub_rep, _ = net_icub.forward(tact_icub)
                bio_rep, _ = net_bio.forward(tact_bio)

            output_icub = net_icub_clr.forward(icub_rep)
            output_bio = net_bio_clr.forward(bio_rep)

            correct_icub += torch.sum(snn.predict.getClass(output_icub) == label).data.item()
            correct_bio += torch.sum(snn.predict.getClass(output_bio) == label).data.item()

            loss_icub = error_icub.numSpikes(output_icub, target)
            loss_bio = error_bio.numSpikes(output_bio, target)

            loss_train_icub += loss_icub.item()
            loss_train_bio += loss_bio.item()

        if epoch%10 == 0:
            print('Epoch: ', epoch, ' --------------------------')
            print('Val loss (icub, bio):', 
                  loss_train_icub/len(val_dataset),
                  loss_train_bio/len(val_dataset))
            print('Val acc (icub, bio):', 
                  correct_icub/len(val_dataset),
                  correct_bio/len(val_dataset))
            
    val_icub_losses.append(loss_train_icub/len(val_dataset))
    val_bio_losses.append(loss_train_bio/len(val_dataset))
    val_icub_accs.append(correct_icub/len(val_dataset))
    val_bio_accs.append(correct_bio/len(val_dataset))
        
        
#     net_icub_clr.eval()
#     net_bio_clr.eval()
#     loss_train_icub = 0
#     loss_train_bio = 0
#     correct_icub = 0
#     correct_bio = 0
#     with torch.no_grad():
#         for i, (tact_icub, tact_bio,  target, label) in enumerate(test_loader):

#             tact_bio = tact_bio.to(device)
#             tact_bio = tact_bio[...,::2]
#             tact_icub = tact_icub.to(device)

#             target = target.to(device)
#             tact_bio = net_bio.get_spike(tact_bio)
#             tact_icub = net_icub.get_spike(tact_icub)

#             with torch.no_grad():
#                 icub_rep, _ = net_icub.forward(tact_icub)
#                 bio_rep, _ = net_bio.forward(tact_bio)

#             output_icub = net_icub_clr.forward(icub_rep)
#             output_bio = net_bio_clr.forward(bio_rep)

#             correct_icub += torch.sum(snn.predict.getClass(output_icub) == label).data.item()
#             correct_bio += torch.sum(snn.predict.getClass(output_bio) == label).data.item()

#             loss_icub = error.numSpikes(output_icub, target)
#             loss_bio = error.numSpikes(output_bio, target)

#             loss_train_icub += loss_icub.item()
#             loss_train_bio += loss_bio.item()

#         if epoch%10 == 0:
#             print('Epoch: ', epoch, ' --------------------------')
#             print('Val loss (icub, bio):', 
#                   loss_train_icub/len(test_dataset),
#                   loss_train_bio/len(test_dataset))
#             print('Val acc (icub, bio):', 
#                   correct_icub/len(test_dataset),
#                   correct_bio/len(test_dataset))
            

    # save bio model on val acc
    if (correct_bio/len(val_dataset) >= max_bio_val_acc):
        print('Saving bio model at ', epoch, ' epoch')
        max_bio_val_acc = correct_bio/len(val_dataset)
        torch.save(net_bio_clr.state_dict(), save_dir_bio)

    # save icub model on val loss
    if (correct_icub/len(val_dataset) >= max_icub_val_acc):
        print('Saving icub model at ', epoch, ' epoch')
        max_icub_val_acc = correct_icub/len(val_dataset)
        torch.save(net_icub_clr.state_dict(), save_dir_icub)
```
