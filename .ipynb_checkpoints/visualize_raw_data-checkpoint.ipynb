```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class ProcessData:
    def __init__(self, path, obj_name, obj_number):
        self.path = path
        self.obj_name = obj_name
        self.obj_number = obj_number
        self.WINDOW_SIZE = 4
        
        # AD-HOC
        if (self.obj_name == 'mat5') & (self.obj_number == 3):
            self.obj_number = 2
        elif (self.obj_name == 'mat19') & (self.obj_number == 6):
            self.obj_number = 7
        
        # prepare file names
#         fpath_info = fdata_path + obj_name + '/' + obj_name + '_' + str(obj_number) + '.txt'
        fpath_bio  = self.path + self.obj_name + '/' + self.obj_name + '_' + str(self.obj_number) + '_bio.csv'
#         fpath_cv   = fdata_path + obj_name + '/' + obj_name + '_' + str(obj_number) + '_cv.csv'
#         fpath_cp   = fdata_path + obj_name + '/' + obj_name + '_' + str(obj_number) + '_cp.csv'
#         fpath_cw   = fdata_path + obj_name + '/' + obj_name + '_' + str(obj_number) + '_cw.csv'
#         fpath_jp   = fdata_path + obj_name + '/' + obj_name + '_' + str(obj_number) + '_jp.csv'
#         fpath_jt   = fdata_path + obj_name + '/' + obj_name + '_' + str(obj_number) + '_jt.csv'
#         fpath_jv   = fdata_path + obj_name + '/' + obj_name + '_' + str(obj_number) + '_jv.csv'
        
        # read files
        df_bio = pd.read_csv(fpath_bio, index_col=False)
        df_bio.timestamp = df_bio.timestamp - df_bio.timestamp[0]
        df_bio_temp = df_bio[ (df_bio.timestamp >= 1) & (df_bio.timestamp <= 1+self.WINDOW_SIZE) ]
        # impute if small data
        if df_bio_temp.shape[0] == 399:
            df_bio_temp = df_bio_temp.append(df_bio.iloc[df_bio_temp.index[-1] + 1])
        self.df_bio = df_bio_temp
    def get_electrode_data(self):
        elctrd_cols = []
        for i in range(1,20):
            elctrd_cols.append('ele' + str(i))
        return self.df_bio[elctrd_cols].values
    def get_pac_data(self):
        pac_vals = []
        for i in range(1,23):
            pac_vals.append('pac' + str(i))
            
        count = 0
        pac_values = np.zeros(self.WINDOW_SIZE*100*22)
        for ind in self.df_bio[pac_vals].index:
            pac_values[count:count+22] = self.df_bio.loc[ind, pac_vals].values
            count += 22
        return pac_values
    def get_timestamp(self):
        return self.df_bio.timestamp.values
```

```python
elctrd_cols = []
for i in range(1,20):
    elctrd_cols.append('ele' + str(i))
```

```python
fpath = '../material_data_Feb/'
```

```python
mat_obj1 = ProcessData(fpath ,'mat15', 1)
mat_obj2 = ProcessData(fpath ,'mat5', 2)
```

```python
mat_obj1.get_electrode_data().shape, mat_obj1.get_timestamp().shape
```

```python
# electrodes
fig, ax = plt.subplots(2)
ax[0].plot(mat_obj1.get_timestamp(), mat_obj1.get_electrode_data())
ax[1].plot(mat_obj2.get_timestamp(), mat_obj2.get_electrode_data())
plt.show()
```
