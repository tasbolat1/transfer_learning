```python
import numpy as np 
import os,sys
import pickle
import matplotlib.pyplot as plt
```

```python
dir = "/home/ruihan/project/transfer_learning/timeseries-clustering-vae/models_and_stats"
filename = "BT19Icub_joint_ae_wrB_0.01_wcB_1_wrI_0.01_wcI_1_wC_1_0.pkl"
with open(os.path.join(dir, filename), "rb") as fin:
    logs = pickle.load(fin)
```

```python
print(len(logs))
print(logs.keys())
```

```python
epoch_train_loss_C = logs['epoch_train_loss_C']
epoch_val_loss_C = logs['epoch_val_loss_C']
print(len(epoch_train_loss_C), len(epoch_val_loss_C))
```

```python
plt.plot(epoch_train_loss_C, label="train")
plt.plot(epoch_val_loss_C, label="val")
plt.xlabel("num epochs")
plt.ylabel("loss_C")
plt.legend()
plt.savefig(os.path.join(dir, "Loss_C_train_val.png"))
```
